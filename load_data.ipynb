```python
try:
    df_clean = df.dropna()
    logging.info(f"Rows after dropping nulls: {len(df_clean)}")

    records = df_clean.to_dict(orient="records")
    print(f"Sample document: {records[0]}")
except Exception as e:
    logging.error(f"Data cleaning failed: {e}")
    print(f"Error: {e}")
```

```python
try:
    collection.delete_many({})
    logging.info("Cleared existing collection")

    batch_size = 1000
    for i in range(0, len(records), batch_size):
        batch = records[i:i+batch_size]
        collection.insert_many(batch)
        logging.info(f"Inserted batch {i//batch_size + 1}")

    print(f"Successfully inserted {collection.count_documents({})} documents")
    logging.info(f"Total documents inserted: {collection.count_documents({})}")
except Exception as e:
    logging.error(f"Insertion failed: {e}")
    print(f"Insertion failed: {e}")
```
